```python
import numpy as np
from sklearn.model_selection import train_test_split as ts
X = np.random.rand(500, 4)
y = np.random.rand(500, 3)

X_train, X_test, y_train, y_test = ts(X, y, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Step 1: Dataset& Dataload

```python
import torch
from torch.utils.data import DataLoader, TensorDataset
```

```python
X_train = torch.FloatTensor(X_train)
X_test  = torch.FloatTensor(X_test)
y_train = torch.FloatTensor(y_train)
y_test = torch.FloatTensor(y_test)
```

```python
X_train
```

```python
train_set = TensorDataset(X_train, y_train)
trainloader = DataLoader(train_set, batch_size=32, shuffle=True)

test_set = TensorDataset(X_test, y_test)
testloader = DataLoader(test_set, batch_size=32)
```

## Step 3 : Training Preparation

### Model|

```python
model = nn.Sequential(nn.Linear(3, 3), 
             nn.ReLU(),
             nn.Linear(3,4), 
             nn.ReLU(),
             nn.Linear(4, 3), 
             nn.Sigmoid()
             )
```

### Creterion

```python
criterion = nn.MSELoss() #model yang diakhiri Linear ->untuk regresi
criterion = nn.NLLLoss() #model yang diakhiri logsoftamax -> multiclass
criterion = nn.BCELoss() #model yang diakhiri sigmoid -> binary clf, multilabelclf
```

### Optimizer

```python
demo.momentum()
```

```python
from torch import optim
optimizer = optim.AdamW(model.parameters(), lr=0.001)
from jcopdl.optim import RAdam, Ranger, RangerLARS, Ralamb
```

### Step 4 : Preparation Training Loop



```python
model.train()
for feature, target in trainloader:
    output = model(feature)
    loss= criterion(output, target)
    loss.backward()
    
    optimizer.step()
    optimizer.zero_gred()
```
